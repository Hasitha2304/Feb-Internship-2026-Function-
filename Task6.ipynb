```python
# Feb Internship Task - 6
# 1. Smart Parking Lot Management System
def manage_parking(capacity, logs):
    parked = 0
    peak = 0
    for entry in logs:
        if entry == "IN":
            parked += 1
        elif entry == "OUT" and parked > 0:
            parked -= 1
        if parked > peak:
            peak = parked
    print("Currently Parked Vehicles:", parked)
    if parked > capacity:
        print("Parking Status: Exceeded Capacity")
    elif parked == capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking(capacity, vehicle_logs)
print()


# 2. Online Food Delivery Time Estimator
def estimate_delivery(distance, traffic, weather):
    base_time = distance * 5   # 5 minutes per km
    # Traffic delay
    if traffic == "High":
        base_time += 10
    elif traffic == "Medium":
        base_time += 5
    # Weather delay
    if weather == "Rainy":
        base_time += 5
    elif weather == "Stormy":
        base_time += 15
    print("Estimated Delivery Time:", base_time, "minutes")
estimate_delivery(8, "High", "Rainy")
print()


# 3. Movie Theatre Seat Occupancy Analyzer
def analyze_occupancy(total_seats, booked_list):
    booked = len(booked_list)
    percentage = (booked / total_seats) * 100
    print("Occupancy:", int(percentage), "%")
    if percentage == 100:
        print("Show Status: Housefull")
    elif percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
booked_seats = [1] * 150
analyze_occupancy(200, booked_seats)
print()


# 4. Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    total = 0
    for value in cpu_readings:
        total += value
    avg = total / len(cpu_readings)
    print("Average CPU Load:", int(avg), "%")
    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
cpu_data = [45, 60, 70, 85, 90]
classify_server_load(cpu_data)
print()


# 5. Smart Classroom Resource Usage Monitor
def monitor_resources(resource_dict):
    overused = []
    for resource in resource_dict:
        if resource_dict[resource] > 8:
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
monitor_resources(resources)
print()


# 6. Online Event Registration Capacity Controller
def manage_registration(capacity, registrations):
    if registrations <= capacity:
        print("Confirmed Registrations:", registrations)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")
    else:
        print("Confirmed Registrations:", capacity)
        print("Waitlisted Users:", registrations - capacity)
        print("Registration Status: Closed")
manage_registration(100, 105)
```
